```python
import pandas as pd
#pd.set_option('display.max_colwidth', None)

# 1. Chargement du fichier
df = pd.read_csv("sortie_questionnaire_traited.csv")

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# 2. Suppression des 7 premières lignes
df = df.iloc[8:].reset_index(drop=True)

# 3. Suppression des lignes de test (par exemple, anciennes lignes 15 et 17)
df = df.drop([8, 10], errors='ignore').reset_index(drop=True)
df.rename(columns={"Pour vous, qu’est-ce qui est important pour votre qualité de vie ? Choisissez 3 images": "dimensions_qdv"}, inplace=True)



colonnes_numeriques = df.select_dtypes(include=['number']).columns.tolist()

df["commune"] = df["Dans quelle commune résidez-vous ? ( A à S)"].combine_first(df["Dans quelle commune résidez-vous ? (T à Z)"])
df.drop(columns=["Dans quelle commune résidez-vous ? ( A à S)", "Dans quelle commune résidez-vous ? (T à Z)"], inplace=True)


# ---------------- Transformation des colonnes ordinales -------------------------
ORDINAL_ENCODINGS = {
    ("Très peu satisfait", "Peu satisfait", "Neutre", "Satisfait", "Très satisfait"): [0, 1, 2, 3, 4],
    ("Très peu entouré", "Peu entouré", "Moyennement entouré", "Bien entouré", "Très bien entouré"): [0, 1, 2, 3, 4],
    ("Très peu impliqué", "Peu impliqué", "Moyenement impliqué", "Impliqué", "Très impliqué"): [0, 1, 2, 3, 4],
    ("Oui", "Non"): [0,1],
    ("Jamais", "Rarement", "Souvent", "Toujours"): [0, 1, 2, 3],
}

# Fonction pour détecter les colonnes ordinales et encoder
def detect_and_encode_ordinal_columns(df, encodings=ORDINAL_ENCODINGS):
    ordinal_cols = []
    encoders = {}

    for col in df.columns:
        unique_values = set(df[col].dropna().unique())
        for label_set, encoding in encodings.items():
            if unique_values.issubset(set(label_set)):
                ordinal_cols.append(col)
                mapping = dict(zip(label_set, encoding))
                encoders[col] = mapping
                df[col + "_encoded"] = df[col].map(mapping)
                break  # Stop at first matching encoding

    return df, ordinal_cols, encoders

# Utilisation
df_encoded, ordinal_columns, encoding_maps = detect_and_encode_ordinal_columns(df)

# Pour visualiser
print("Colonnes détectées comme ordinales :", ordinal_columns)
print("Exemple de mapping :", encoding_maps)


#--------------- Calcul du pourcentage de chaque dimension de la qualité de vie par commune -----------------
# explosion des réponses multiples
df_dimensions = df[["commune", "dimensions_qdv"]].dropna().copy()
df_dimensions["dimensions_qdv"] = df_dimensions["dimensions_qdv"].str.split(";")
df_exploded = df_dimensions.explode("dimensions_qdv")

# nettoyage
df_exploded["dimensions_qdv"] = df_exploded["dimensions_qdv"].str.strip()

#  calcul du nombre total de répondants par commune
respondants_per_commune = df.groupby("commune").size().rename("total_respondants")

# Étape 4 : comptage des dimensions par commune
dimension_counts = (
    df_exploded.groupby(["commune", "dimensions_qdv"]).size().rename("count").reset_index()
)

# Étape 5 : merge pour calculer les pourcentages
dimension_counts = dimension_counts.merge(respondants_per_commune, on="commune")
dimension_counts["percentage"] = (
    dimension_counts["count"] / dimension_counts["total_respondants"] * 100
).round(2)
print(df_dimensions.head())


#------------------ Création d'un dataset regroupant les verbatims par commune ---------------
print(df.columns.tolist())

# Renommer les colonnes des verbatims
df.rename(columns={
    "Justifiez brièvement le choix de votre première\xa0image (5 mots).": "verbatim_1",
    "Justifiez brièvement le choix de votre deuxième image (5 mots).": "verbatim_2",
    "Justifiez brièvement le choix de votre troisième image (5 mots).": "verbatim_3",
}, inplace=True)

# Garder uniquement les colonnes utiles
verbatims_df = df[["commune", "verbatim_1", "verbatim_2", "verbatim_3"]].copy()

# Transformer en format long (1 ligne = 1 verbatim)
verbatims_long = verbatims_df.melt(id_vars="commune", value_vars=["verbatim_1", "verbatim_2", "verbatim_3"],
                                    var_name="verbatim_source", value_name="verbatim")

#  Nettoyage (on garde les verbatims non vides)
verbatims_long["verbatim"] = verbatims_long["verbatim"].astype(str).str.strip()
verbatims_long = verbatims_long[verbatims_long["verbatim"].notnull() & (verbatims_long["verbatim"] != "")]

# Étape 5 : Grouper par commune, et concaténer les verbatims
verbatims_by_commune = (
    verbatims_long.groupby("commune")["verbatim"]
    .apply(lambda x: list(x))
    .reset_index()
)


#------------------------- Traitement des données numériques - moyennes par commune


df_copy = df.copy()

# Identifier les colonnes numériques (excluant 'commune' si c'est une chaîne)
numerical_cols = df_copy.select_dtypes(include=['number']).columns.tolist()

# S'assurer que 'commune' est dans le dataframe et ne soit pas dans numerical_cols
if 'commune' in numerical_cols:
    numerical_cols.remove('commune')

# Ajouter 'commune' à la liste des colonnes à garder (pour grouper)
cols_to_keep = ['commune'] + numerical_cols

# Extraire uniquement ces colonnes
df_numerical = df_copy[cols_to_keep]
print(df_numerical.columns.tolist())

# Calculer la moyenne par commune
df_mean_by_commune = df_numerical.groupby('commune').mean().reset_index()

# Prints
"""print(df_mean_by_commune.head())

print(verbatims_by_commune.head())

print("!!!!!!!! IMPORTANT !!!!!!")

print(verbatims_by_commune.columns.tolist())
print(df_mean_by_commune.columns.tolist())
print(dimension_counts.columns.tolist())"""


"""
res = df.columns
print(res)
print(dimension_counts.head(25))
"""

""" ----------------------------------- Génération de fichiers RAG-friendly !!---------------"""

import os
import pandas as pd

# === CONFIGURATION DES DOSSIERS DE SORTIE ===
output_dir_quanti = "rag_quanti"
output_dir_be = "rag_be"
os.makedirs(output_dir_quanti, exist_ok=True)
os.makedirs(output_dir_be, exist_ok=True)

# === RENOMMAGE POUR LISIBILITÉ (FACULTATIF MAIS UTILE) ===
df_mean_by_commune = df_mean_by_commune.rename(columns={
    'Sur une échelle de 1 à 5, pourriez-vous estimer à quel point vous êtes heureux ces derniers temps ?': 'Score bonheur',
    'Sur une échelle de 1 à 5,\xa0pourriez-vous évaluer votre qualité de vie ces derniers temps ?': 'Score qualité de vie',
    "Sur une échelle de 1 à 5, pourriez-vous évaluer votre confiance en l'avenir ?": 'Score confiance avenir',
    "Les services de transports_encoded": "Transports",
    "L'accès à l'éducation_encoded": "Éducation",
    "La couverture des réseaux téléphoniques_encoded": "Réseaux téléphoniques",
    "Les institutions étatiques (niveau de confiance)_encoded": "Institutions",
    "Le tourisme (ressenti localement)_encoded": "Tourisme",
    "La sécurité_encoded": "Sécurité",
    "L'offre de santé _encoded": "Santé",
    "Votre situation professionnelle_encoded": "Situation pro",
    "Vos revenus_encoded": "Revenus",
    "La répartition de votre temps entre travail et temps personnel_encoded": "Temps travail/perso",
    "Votre logement_encoded": "Logement",
    "L'offre de services autour de chez vous_encoded": "Services locaux",
    "Votre accès à la culture_encoded": "Culture",
    "Vous sentez-vous bien entouré ?_encoded": "Soutien social",
    "Êtes-vous impliqué dans des activités associatives ?_encoded": "Vie associative"
})

# === FONCTIONS POUR CRÉER LES TEXTES ===

def generate_quanti_text(commune, df_mean):
    row = df_mean[df_mean['commune'] == commune]
    if row.empty:
        return f"Commune : {commune}\n\nAucune donnée quantitative disponible."

    row = row.iloc[0]
    text = f"Commune : {commune}\n\nMoyennes des indicateurs de qualité de vie :\n"
    for col in df_mean.columns:
        if col not in ['commune', 'ID']:
            text += f"- {col} : {row[col]:.2f}\n"
    return text

def generate_be_text(commune, df_dims, df_verbatims):
    dims = df_dims[df_dims['commune'] == commune]
    verb_row = df_verbatims[df_verbatims['commune'] == commune]

    text = f"Commune : {commune}\n\nDimensions du bien-être identifiées :\n"

    if dims.empty:
        text += "- Aucune dimension renseignée.\n"
    else:
        for _, row in dims.iterrows():
            dim = row['dimensions_qdv']
            pct = row['percentage']
            text += f"- {dim} : sélectionnée par {pct:.1f}% des répondants\n"

    text += "\nVerbatims associés :\n"
    if verb_row.empty or not verb_row.iloc[0]['verbatim']:
        text += "- Aucun verbatim disponible.\n"
    else:
        for i, v in enumerate(verb_row.iloc[0]['verbatim'], 1):
            text += f"{i}. {v}\n"
    return text

# === LISTE DES COMMUNES ===
communes = set(df_mean_by_commune['commune']).union(dimension_counts['commune']).union(verbatims_by_commune['commune'])

# === GÉNÉRATION DES FICHIERS ===
for commune in communes:
    safe_filename = commune.replace(" ", "_").replace("/", "-")

    # Texte quantitatif
    quanti_text = generate_quanti_text(commune, df_mean_by_commune)
    with open(os.path.join(output_dir_quanti, f"{safe_filename}.txt"), "w", encoding="utf-8") as f:
        f.write(quanti_text)

    # Texte dimensions + verbatims
    be_text = generate_be_text(commune, dimension_counts, verbatims_by_commune)
    with open(os.path.join(output_dir_be, f"{safe_filename}.txt"), "w", encoding="utf-8") as f:
        f.write(be_text)

print("✅ Tous les fichiers RAG-friendly ont été générés par commune.")

# Ajout du nombre de répondants à df_mean_by_commune
df_mean_by_commune = df_mean_by_commune.merge(respondants_per_commune.reset_index(), on="commune")


df_mean_by_commune.to_csv("df_mean_by_commune.csv", index=False, encoding="utf-8")
verbatims_by_commune.to_csv("verbatims_by_commune.csv", index=False, encoding="utf-8")
dimension_counts.to_csv("dimension_counts.csv", index=False, encoding="utf-8")
```

```python
import json
from rdflib import Graph

# ---------
# Fonctions utilitaires
# ---------
def stringify(node, graph):
    """Convertit un noeud RDF en version lisible"""
    node_str = str(node)

    # Si c'est un blank node
    if node_str.startswith("_:"):
        return "BlankNode"

    # Essaie de raccourcir avec le préfixe
    try:
        prefix, namespace, name = graph.namespace_manager.compute_qname(node_str)
        return f"{prefix}:{name}"
    except Exception:
        return node_str

# ---------
# Charger le graphe
# ---------
g = Graph()
g.parse("ontology_be_2010.ttl", format="turtle")  # change le nom si besoin

# ---------
# Extraire les triplets
# ---------
triplets = []
for subj, pred, obj in g:
    triplets.append({
        "sujet": stringify(subj, g),
        "predicat": stringify(pred, g),
        "objet": stringify(obj, g)
    })

# ---------
# Afficher en JSON lisible
# ---------
print(json.dumps(triplets, indent=4, ensure_ascii=False))
```

```python
import re
import os
import pickle
import chromadb
from chromadb.config import Settings
from sentence_transformers import SentenceTransformer
from tqdm import tqdm
import nltk
nltk.download("punkt")
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize

# --- Fonction de découpage des textes en chunks de ~150 mots ---
def chunk_text(text, source_type="default"):
    max_words = {
        "entretien": 80,
        "wiki": 150,
        "quanti": 50,
        "be_verbatims": 60,
        "csv_text": 100,
        "default": 100
    }.get(source_type, 100)

    sentences = sent_tokenize(text)
    chunks = []
    current_chunk = []
    current_len = 0

    for sentence in sentences:
        words = sentence.split()
        if current_len + len(words) > max_words:
            chunks.append(" ".join(current_chunk))
            current_chunk = words
            current_len = len(words)
        else:
            current_chunk.extend(words)
            current_len += len(words)
    if current_chunk:
        chunks.append(" ".join(current_chunk))
    return chunks


# --- Initialisation ChromaDB ---
chroma_client = chromadb.PersistentClient(path="./chroma_txt")
collection = chroma_client.get_or_create_collection(name="communes_corses_txt")

# --- Vérification du cache ---
if os.path.exists("embeddings.pkl"):
    print("✅ Chargement des embeddings depuis le cache...")
    with open("embeddings.pkl", "rb") as f:
        data = pickle.load(f)
    documents = data["documents"]
    embeddings = data["embeddings"]
    metadatas = data["metadatas"]
    ids = data["ids"]

else:
    print("🚀 Aucune cache trouvée, génération des embeddings...")

    documents = []
    metadatas = []
    ids = []
    chunk_counter = 0

    # --- Traitement du fichier CSV texte ---
    csv_text_path = "/content/sample_data/communes_text.txt"
    with open(csv_text_path, "r", encoding="utf-8") as f:
        lines = f.readlines()
        lines = lines[:10]

    for i, line in enumerate(lines):
        line = line.strip()
        if not line:
            continue
        match = re.match(r"La commune de ([^ ]+)", line)
        commune_name = match.group(1) if match else f"commune_{i}"
        chunks = chunk_text(line, source_type="wiki")
        for j, chunk in enumerate(chunks):
            documents.append(chunk)
            metadatas.append({
                "nom": commune_name,
                "source": "csv_text",
                "chunk": j
            })
            ids.append(f"csv_text_commune_{i}_chunk_{j}")
            chunk_counter += 1

     #--- Traitement des entretiens semi-directifs ---
    entretiens_path = "/content/sample_data/entretiens_lea.txt"
    with open(entretiens_path, "r", encoding="utf-8") as f:
      raw_entretiens = f.read()

    # On découpe chaque bloc d'entretien (séparé par des lignes de tirets)
    blocs_entretiens = re.split(r"\n[-]{3,}\n", raw_entretiens.strip())

    for bloc in blocs_entretiens:
        # On cherche le header : ### Commune : Nom | Entretien X
        match = re.match(r"###\s*Commune\s*:\s*(.+?)\s*\|\s*Entretien\s*(\d+)\s*\n+(.+)", bloc, re.DOTALL)
        if match:
            nom_commune, num_entretien, texte_entretien = match.groups()

            # On découpe le texte complet de l’entretien en chunks (~150 mots)
            chunks = chunk_text(texte_entretien.strip(), source_type="wiki")
            for j, chunk in enumerate(chunks):
                documents.append(chunk)
                metadatas.append({
                    "nom": nom_commune.strip(),
                    "entretien": int(num_entretien),
                    "source": "entretien",
                    "chunk": j
                })
                ids.append(f"entretien_{nom_commune.strip().replace(' ', '_')}_{num_entretien}_chunk_{j}")
                chunk_counter += 1

                # 🖨️ Impression de contrôle
                print(f"\n📍 Commune: {nom_commune} | Entretien: {num_entretien} | Chunk {j}")
                print(f"➡️ Aperçu : {chunk[:200]}...")

    else:
        print(f"❗ Bloc d'entretien ignoré (non conforme) :\n{bloc[:100]}...")

    # --- Traitement du fichier wiki ---
    txt_path = "communes_corses_wiki.txt"
    with open(txt_path, "r", encoding="utf-8") as f:
        raw_text = f.read()

    blocs = re.split(r"\n---+\n", raw_text.strip())
    blocs = blocs[:10]  # 👈 limite aux 10 premiers blocs wiki

    for i, bloc in enumerate(blocs):
        match = re.search(r"### (.+?)\n\nRésumé : (.+?)\n\nDescription : (.+)", bloc, re.DOTALL)
        if match:
            nom, resume, description = match.groups()
            full_text = f"{resume.strip()}\n\n{description.strip()}"
            chunks = chunk_text(full_text, source_type = "wiki")
            for j, chunk in enumerate(chunks):
                documents.append(chunk)
                metadatas.append({
                    "nom": nom.strip(),
                    "résumé": resume.strip(),
                    "source": "wiki",
                    "chunk": j
                })
                ids.append(f"commune_{i}_chunk_{j}")
                chunk_counter += 1
        else:
            print(f"❗ Bloc ignoré (non conforme) :\n{bloc[:100]}...")

# --- Intégration des fichiers rag_quanti et rag_be (1 chunk par fichier) ---
    import glob

    def add_rag_fichiers_txt(folder_path, source_label):
        txt_files = glob.glob(os.path.join(folder_path, "*.txt"))
        for path in txt_files:
            with open(path, "r", encoding="utf-8") as f:
                text = f.read().strip()
                if not text:
                    continue

            filename = os.path.basename(path)
            nom_commune = filename.replace(".txt", "").replace("_", " ").strip()

            documents.append(text)
            metadatas.append({
                "nom": nom_commune,
                "source": source_label,
                "chunk": 0
            })
            ids.append(f"{source_label}_{nom_commune.replace(' ', '_')}_chunk_0")

            print(f"📄 Ajouté : {filename} | source={source_label}")

    add_rag_fichiers_txt("rag_quanti", "quanti")
    add_rag_fichiers_txt("rag_be", "be_verbatims")



    # --- Embedding par batches ---
    model = SentenceTransformer("intfloat/e5-base-v2")
    documents_prefixed = [f"passage: {doc}" for doc in documents]

    embeddings = []
    batch_size = 128
    for i in tqdm(range(0, len(documents_prefixed), batch_size), desc="🔄 Encodage des chunks"):
        batch = documents_prefixed[i:i + batch_size]
        batch_embeddings = model.encode(batch).tolist()
        embeddings.extend(batch_embeddings)

    # --- Sauvegarde des embeddings en cache ---
    with open("embeddings.pkl", "wb") as f:
        pickle.dump({
            "documents": documents,
            "embeddings": embeddings,
            "metadatas": metadatas,
            "ids": ids
        }, f)
    print(f"✅ Embeddings sauvegardés ({chunk_counter} chunks) dans embeddings.pkl")

# --- Indexation par batches ---
batch_size = 5000
for i in tqdm(range(0, len(documents), batch_size), desc="🔄 Indexation des chunks"):
    collection.add(
        documents=documents[i : i + batch_size],
        embeddings=embeddings[i : i + batch_size],
        metadatas=metadatas[i : i + batch_size],
        ids=ids[i : i + batch_size],
    )

print(f"✅ Indexation réussie ! {len(documents)} chunks ajoutés.")
```

```python
from sentence_transformers import SentenceTransformer
import chromadb

# --- Connexion à la base persistée ---
chroma_client = chromadb.PersistentClient(path="./chroma_txt")
collection = chroma_client.get_collection(name="communes_corses_txt")

# --- Modèle pour encoder la requête ---
model = SentenceTransformer("intfloat/e5-base-v2")

# --- Question en langage naturel ---
question = "query: Combien y a-t-il de médecins à Afa ?"

# --- Encodage en vecteur ---
query_vector = model.encode([question]).tolist()

# --- Recherche dans l'index ---
results = collection.query(
    query_embeddings=query_vector,
    n_results=10  # Nombre de réponses à retourner
)

# --- Affichage des résultats ---
for doc, metadata in zip(results["documents"][0], results["metadatas"][0]):
    print(f"🔹 Commune : {metadata['nom']}")
    print(f"Résumé : {metadata['résumé']}")
    print(f"---\n{doc[:300]}...\n")

from collections import defaultdict

grouped = defaultdict(list)

# Regrouper les résultats par nom de commune
for doc, metadata, dist in zip(results["documents"][0], results["metadatas"][0], results["distances"][0]):
    grouped[metadata["nom"]].append((dist, doc, metadata))

# Trouver la commune avec le chunk le plus pertinent (distance minimale)
best_match = min(
    ((min(chunks, key=lambda x: x[0])[0], nom, chunks) for nom, chunks in grouped.items()),
    key=lambda x: x[0]
)

# Afficher la meilleure commune
print(f"\n🏆 Commune la plus pertinente : {best_match[1]} (distance {best_match[0]:.4f})")
print("🔍 Meilleur chunk :")
print(best_match[2][0][1][:500] + "...\n")  # Affiche le texte
print(f"Résumé : {best_match[2][0][2]['résumé']}")
```

```python
import openai
import chromadb
from sentence_transformers import SentenceTransformer
from collections import defaultdict

openai.api_key = ""

def ask_rag(question, n_chunks=5, model_name="gpt-3.5-turbo"):
    # 1. Encodage de la requête
    embed_model = SentenceTransformer("intfloat/e5-base-v2")
    query_embedding = embed_model.encode([f"query: {question}"]).tolist()

    # 2. Recherche dans la base Chroma
    results = collection.query(
        query_embeddings=query_embedding,
        n_results=n_chunks
    )

    retrieved_docs = results["documents"][0]
    retrieved_context = "\n\n".join(retrieved_docs)

    # 3. Création du prompt
    prompt = f"""Tu es un conseiller municipal. Ton but est de donner des informations sur la qualité de vie dans les communes Corses, pour guider les politiques publiques, en te basant unniquement sur les informations suivantes :

{retrieved_context}

Question : {question}
Réponse :
"""

    # 4. Appel au LLM
    response = openai.ChatCompletion.create(
        model=model_name,
        messages=[
            {"role": "system", "content": "Tu es un assistant utile et factuel. Ne réponds qu'avec les informations données. Commence toutes tes phrases en disant si dieu le veut"},
            {"role": "user", "content": prompt}
        ]
    )

    answer = response["choices"][0]["message"]["content"]
    return answer

reponse = ask_rag("Quels sont les thèmes de prédilection pour la qualité de vie à Corte ?")

print("🧠 Réponse :", reponse)
```
